```python
import pandas as pd
from datetime import datetime

df = pd.read_csv('tarkov-tools-price-data.csv')
names = pd.read_csv('items_by_id.csv')

df = pd.merge(df,names,how='left',on='item_id')

df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
oli_log = df.loc[df['name'].str.contains('OLI logistics department office key')]
ultra = df.loc[df['name'].str.contains('ULTRA')]
nec = df.loc[df['name'].str.contains('NecrusPharm pharmacy key')]
kiba_outer = df.loc[df['name'].str.contains('Kiba Arms outer door key')]
kiba_inner = df.loc[df['name'].str.contains('Kiba Arms inner grate door key')]
emercom = df.loc[df['name'].str.contains('EMERCOM medical unit key')]
oli_cash = df.loc[df['name'].str.contains('OLI cash register key')]
idea_cash = df.loc[df['name'].str.contains('IDEA cash register key')]
goshan_cash = df.loc[df['name'].str.contains('Goshan cash register key')]
ob12 = df.loc[df['name'].str.contains('21WS')]
ob11 = df.loc[df['name'].str.contains('11SR')]
```

```python
interchange = oli_log.append(ultra)
interchange = interchange.append(nec)
interchange = interchange.append(kiba_outer)
interchange = interchange.append(kiba_inner)
interchange = interchange.append(emercom)
interchange = interchange.append(oli_cash)
interchange = interchange.append(idea_cash)
interchange = interchange.append(goshan_cash)
interchange = interchange.append(ob12)
interchange = interchange.append(ob11)
interchange
```

```python
interchange.to_csv('interchange.csv')
```
